# OPEN CV

```python
import cv2
```

```python
img = cv2.imread('C:/Users/aaana/Desktop/rock.jpg')
```

```python
type(img)   #type of the image 
```

```python
img.shape  #shape of the image
```

```python
plt.imshow(img)
```

```python
img_fix = cv2.cvtColor(img,
                      cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_fix)
```

##### Scale it to <b>Gray</b> and check the <b>Shape</b>

```python
img_gray = cv2.imread('C:/Users/aaana/Desktop/rock.jpg',
                     cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap='gray')
```

```python
img_new = cv2.resize(img_fix,
                    (1000, 700))
```

```python
plt.imshow(img_new)
```

### Resize with Ratio

```python
width_ratio = 0.5
height_ratio = 0.5
```

```python
img2 = cv2.resize(img_fix,
                 (0,0),
                 img_fix,
                 width_ratio,
                 height_ratio)
```

```python
plt.imshow(img2)
```

```python
img2.shape
```

### Flip on Horizontal Axis

```python
img_3 = cv2.flip(img_fix, 0)
plt.imshow(img_3)
```

### Flip on Vertical Axis

```python
img_3 = cv2.flip(img_fix, 1)
plt.imshow(img_3)
```

### Flip on Horizontal and on Vertical Axis

```python
img_3 = cv2.flip(img_fix,-1)
plt.imshow(img_3)
```
