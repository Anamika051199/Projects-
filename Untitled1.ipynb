```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import time
```

```python
data = pd.read_csv('data.csv')
data.head()
```

```python
col = data.columns
```

```python
print(col)
```

```python
y = data.diagnosis
drop_cols = ['Unnamed: 32','id','diagnosis']
x = data.drop(drop_cols, axis=1)
x.head()
```

```python
ax = sns.countplot(y, label ="count")
B, M = y.value_counts()
print("Number of Begnin tumours", B)
print("Number of Malignant tumours", M)
```

```python
x.describe()
```

```python
data = x
data_std = (data-data.mean())/data.std()
data = pd.concat([y, data_std.iloc[: ,0:10]],axis=1)
data = pd.melt(data,id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x='features',y='value',hue='diagnosis', data=data, split=True)
plt.xticks(rotation=45)
```

```python
sns.boxplot(x= 'features', y='value',hue='diagnosis', data=data)
plt.xticks(rotation=45)
```

```python
sns.jointplot(x.loc[:, 'concavity_worst'],
             x.loc[:, 'concave points_worst'],
             kind = 'regg',
             color = '#ce1414')
```

```python
sns.set(style='whitegrid', palette='muted')
data = x
data_std = (data-data.mean())/data.std()
data = pd.concat([y, data_std.iloc[: ,0:10]],axis=1)
data = pd.melt(data,id_vars='diagnosis',
              var_name='features',
              value_name='value')
plt.figure(figsize=(10,10))
sns.swarmplot(x='features',y='value',hue='diagnosis', data=data, split=True)
plt.xticks(rotation=45)
```
